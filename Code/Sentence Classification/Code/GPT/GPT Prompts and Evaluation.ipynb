```python
# Save the prompts and sentences for ones over the token limit by creating multiple files 
# VERSION: GPT-3.5, Justify Answer
# GPT-3.5-Turbo Max Tokens: 4906 
# Our max tokens: if using API would have to be significantly less to account for response 
import os


def save_to_file(text, filepath):
    with open(filepath, "w") as f:
        f.write(text)


def save_dict_strings(dictionary, model_name, directory, prompt):
    """
    Saves each value of a dictionary as a text file in a separate subdirectory for each key. The text will be pre-pended with a given prompt.
    If the text value has a token count exceeding 4095, it will be split into multiple files, each with token count <= 4095.

    Args:
    dictionary (dict): Dictionary with string values to be saved as files.
    model_name (str): The name of the GPT model to be used for token counting.
    directory (str): The directory where the files will be saved.
    prompt (str): The prompt to be pre-pended to each file.
    """
    for key, string in dictionary.items():
        if num_tokens_from_string(string + prompt, model_name) > 4095:
          key_directory = os.path.join(directory, key)
          os.makedirs(key_directory, exist_ok=True)
          sentences = string.split(". ")
          file_counter = 0
          while sentences:
              current_chunk = prompt
              while sentences and num_tokens_from_string(current_chunk + sentences[0] + ". ", model_name) <= 4095:
                  current_chunk += sentences.pop(0) + ". "
              filepath = os.path.join(key_directory, f"{key}_{file_counter}.txt")
              save_to_file(current_chunk, filepath)
              print("Saved to {}".format(filepath))
              file_counter += 1


gpt_save_folder = "/content/drive/MyDrive/Colab Notebooks/NLP - Lab/Optic Nerve Regeneration/Code/Sentence Classification/Output/GPT Prompts/Justify Answer/"
model_name = "gpt-3.5-turbo"

save_dict_strings(labeled_sentences, model_name, gpt_save_folder, gpt_justify_answer_prompt)
```
