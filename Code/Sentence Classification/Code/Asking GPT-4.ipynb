# Sentences to Give to GPT



*   Generated MAG sentences 
* Generate np-1 sentences since that is one our BERT model got wrong (labelled as inhibitor when it should be a promoter) 




```python
# pad the molecule with spaces so it doesn't search in-between words
molecule_double_spaced = ' np-1 ' 
trainable_sentences_for_molecule = [sentence for sentence in all_docs_1D if molecule_double_spaced in sentence]
len(trainable_sentences_for_molecule)
```

```python
trainable_sentences_for_molecule
with open("/content/drive/MyDrive/Colab Notebooks/NLP - Lab/GraphSage/Output/trainable_sentences_{}.txt".format(molecule_double_spaced.strip()), 'w') as f: 
    for s in trainable_sentences_for_molecule:
        f.write(s)
```

```python
# We will now try to mask the molecule by labelling it as '[molecule]' in the sentence to ensure that GPT isn't cheating and using the 
# name when coming to its conclusion 

# NOTE: When we don't mask then GPT can also use other instances of the molecule in the sentence that aren't padded with spaces to help 
# get to its conclusion but this version it can't 

trainable_sentences_for_molecul_masked = [sentence.replace(molecule_double_spaced, ' [molecule] ') for sentence in all_docs_1D if molecule_double_spaced in sentence]
with open("/content/drive/MyDrive/Colab Notebooks/NLP - Lab/GraphSage/Output/trainable_sentences_masked_{}.txt".format(molecule_double_spaced.strip()), 'w') as f: 
    for s in trainable_sentences_for_molecul_masked:
        f.write(s)
```
